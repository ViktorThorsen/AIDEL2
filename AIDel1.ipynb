#### 3a. Vad är syftet med att dela upp data i träningsdata, valideringsdata och testdata?

Syftet med att dela upp data i träningsdata, valideringsdata och testdata är att säkerställa att modellen inte bara lär sig utantill (överanpassar), utan också kan generalisera till ny, osedd data.
Träningsdata används för att lära modellen – det är här den justerar sina parametrar.
Valideringsdata används för att jämföra olika modeller och ta fram vilken som presterar bäst.
Testdata används i slutet och testar hur bra modellen fungerar på ny data.

#### 3b. Vad är k-delad korsvalidering för något?

#### 3d. Vad är en hyperparameter för något? Vad är en parameter för något?

En hyperparameter är en inställning som du själv bestämmer innan träningen av en maskininlärningsmodell startar. De styr hur modellen lär sig.
En parameter är något som modellen själv räknar ut under träningen. Det är modellens inlärda värden, till exempel vikterna i en linjär modell och interceptet (skärningspunkten).

#### 3e. Vad är grid search

GridSearch ett sätt att automatiskt prova många hyperparameterinställningar och hitta den som gör modellen bäst.
Man skapar ett rutnät (grid) med möjliga värden på olika hyperparametrar.
Sedan söks (search) hela rutnätet igenom genom att testa varje kombination.
För varje kombination utvärderas modellen med cross-validation för att se hur bra den fungerar.
Till slut väljs den kombination som ger bäst resultat.

När refit=True så tränas modellen om på hela datasetet med de bästa hyperparametrarna som hittades.

#### 3f. Vad är kategorisk data och hur hanteras det?

Kategorisk data är data som består av olika kategorier istället för tal, till exempel färger eller bilmärken. För att kunna använda sådan data i maskininlärningsmodeller behöver den göras om till numerisk form.

Nominal data är kategorier som inte har någon inbördes rangordning, till exempel olika färger. 
Ordinal data är kategorier som har en naturlig rangordning, till exempel utbildningsnivå där grundskola kommer före gymnasie och gymnasie före högskola.

För att hantera kategorisk data kan man använda olika metoder. Ordinal encoding används för ordinal data där varje kategori får ett numeriskt värde i stigande ordning, till exempel låg = 1, medel = 2 och hög = 3. One-hot-encoding används ofta för nominal data där varje kategori blir en egen kolumn som får värdet 0 eller 1, till exempel bilmodell Volvo blir [1,0,0] och bilmodell BMW blir [0,1,0]. Dummy-variable-encoding fungerar på liknande sätt som one-hot men man tar bort en kolumn, till exempel om man har tre färger så skapar man bara två kolumner

#### 3g. Vad är feature engineering för något?

Feature engineering innebär att man väljer ut vilka variabler (features) som ska användas i modellen och hur de ska representeras. Det kan handla om att välja bort irrelevanta variabler men också att skapa nya eller omvandla befintliga, till exempel normalisera värden eller slå ihop variabler. Det är viktigt eftersom en modell bara kan bli så bra som den data den får. Om modellen får dåliga eller irrelevanta variabler kommer den att ge dåliga resultat.

#### 3h. Vad menas med principle of parsimony?

Det innebär att man inte ska göra modellen mer komplicerad än nödvändigt. En enkel modell som fångar de viktiga sambanden är ofta bättre än en väldigt komplex modell som riskerar att överanpassa (overfitting).

2. Förklara utvärderingsmåtten RMSE, MSE och MAE.

RMSE (Root Mean Squared Error) = används ofta och visar skillnaden mellan det förutsagda värdet och det riktiga värdet. Man tar kvadraten på skillnaderna, räknar ut medelvärdet och sen roten ur. Ett lägre värde betyder en bättre modell.

MSE (Mean Squared Error) = funkar nästan som RMSE, men man tar inte roten ur i slutet. Det gör att värdena blir lite större och inte lika lätt att tolka i samma enhet, men det går snabbare att räkna ut.

MAE (Mean Absolute Error) = här tar man inte kvadraten på skillnaderna, utan bara det absoluta felet (alltså hur långt ifrån man är, oavsett plus eller minus). Sen tar man medelvärdet av alla dessa. Det gör att man får ett mer rakt mått på det genomsnittliga felet, och det påverkas inte lika mycket av extremvärden som MSE och RMSE gör.

#### 2. Förklara hur OvR- och OvO-algoritmerna fungerar.

##### One-vs-Rest (OvR)

Man har flera klasser, säg A, B, C.

OvR tränar en klassificerare för varje klass mot “alla andra”.

Klassificerare 1: A vs (B och C)

Klassificerare 2: B vs (A och C)

Klassificerare 3: C vs (A och B)

När man ska förutsäga för en ny datapunkt kör man den genom alla klassificerare, och väljer den klass som gav högst sannolikhet/poäng.

##### One-vs-One (OvO)

Här tränar man en klassificerare för varje par av klasser.

Med A, B, C får man:

Klassificerare 1: A vs B

Klassificerare 2: A vs C

Klassificerare 3: B vs C

När man ska förutsäga körs datapunkten genom alla klassificerare, och varje klass får “röster”. Den klass som får flest röster blir slutresultatet.

#### 3. Förklara följande utvärderingsmått:

#### a) Confusion matrix

En tabell som visar hur många förutsägelser som var rätt eller fel, uppdelat per klass.

#### b) Accuracy

#### c) Precision

Hur många av de som modellen förutspådde som positiva var faktiskt positiva.

#### d) Recall

Hur många av de faktiskt positiva hittade modellen.

#### e) F1-score

Ett balanserat mått som kombinerar precision och recall i ett enda tal.

#### f) ROC-kurvan

ROC = Receiver Operating Characteristic.

Man varierar tröskeln för vad modellen räknar som positiv.

På y-axeln: True Positive Rate (Recall).

På x-axeln: False Positive Rate 

Kurvan visar balansen mellan att fånga positiva och risken att skapa falsklarm.

Att om vi maxar precision så kommer recall sjunka och vi kommer missa fler positiva.

i rättsväsendet kan det vara bra att ha hög precision eftersom vi inte vill dömma någon oskyldig så är det viktigt att de positiva stämmer.

# Kapitel 10 - Chattbottar

#### 1. Vad är prompt engineering?

Prompt enginering innebär att förfina och justera prompter till LLM:er så att man får det svar man vill ha, samtidigt som det använder så lite tokens och datorkraft som möjligt.

#### 2. Vad är RAG?

RAG (Retrieval-Augmented Generation) är en teknik som ger AI:n tillgång till extern, uppdaterad information utanför dess förtränade kunskap.

#### 3. Vad är chunking och embeddings?

Chunking innebär att man delar in lång text i delar, på så sätt kan man få tag i de delar som är relevanta av tex en större pdf. Detta gör att man kan dra ner kostnaden och underlätta för LLMen istället för att behöva använda hela pdfen i varje call.

Embeddings är processen där textbitarna (dina chunks) översätts till listor med siffror, så kallade vektorer.

#### 4. Förklara översiktligt hur man kan evaluera en chattbot.

Man evaluerar en chatbot genom att kolla på både teknisk precision (typ RAG-metrics) och hur användarna faktiskt upplever hjälpsamheten. Det handlar om att mäta att den inte hittar på grejer, men också att den svarar snabbt och i rätt tonläge för att vara användbar på riktigt.

#### 5. Din kollega frågar dig vad ELIZA och Turingtestet är. Läs nedanstående två länkar innan du besvarar frågan.

ELIZA skapades på 60-talet och var en av de första chattbotarna som simulerade mänsklig konversation genom att imitera en psykoterapeut. Den använde enkla mönster för att spegla användarens egna ord som frågor, vilket skapade en illusion av förståelse trots att den saknade verklig intelligens.

Turingtestet Detta är ett klassiskt test från 1950 för att avgöra om en maskin uppvisar mänsklig intelligens. Om en människa chattar med en maskin och en annan människa utan att kunna avgöra vem som är vem, anses maskinen ha klarat testet.
